```python
class Carta():
    def __init__(self,idObjetc ,idClase,nombre,tipo):
        self.idObjetc = idObjetc
        self.idClase = idClase
        self.nombre = nombre
        self.tipo = tipo
        
    def __str__(self):
        return "Carta: "+ str(self.idObjetc)+" / " + "Clase: " + str(self.idClase)+" / " + "Nombre: " +self.nombre+" / " + "Tipo: " +self.tipo
```

```python
class Personaje():
    def __init__(self,idPersonaje,nombre,vidas):
        self.idPersonaje = idPersonaje
        self.nombre = nombre
        self.vidas = vidas
    def __str__(self):
        return "idPersonaje: "+ str(self.idPersonaje)+" / " + "nombre: " + str(self.nombre)+" / " + "vidas: " +self.vidas
```
